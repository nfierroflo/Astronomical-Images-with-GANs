```python
num_img=5000

gen_1 = Generator(z_dim).to(device)
gen_1.load_state_dict(torch.load('dcgan'))
gen = Generator(z_dim).to(device)
gen.load_state_dict(torch.load('wgangp'))
fake_noise = get_noise(num_img, z_dim, device=device)
fake_dcgan = gen_1(fake_noise)
fake_wgangp= gen(fake_noise)
```

```python
from torchvision.utils import make_grid

def show_tensor_images(image_tensor, num_images=25, num_channels=3, num_columns=3):
    image_tensor = (image_tensor + 1) / 2
    image_unflat = image_tensor.detach().cpu()

    fig, axes = plt.subplots(num_images // num_channels, num_columns, figsize=(12, 8))
    
    for i in range(num_images):
        row = i // num_channels
        col = i % num_channels
        channel_images = image_unflat[i, :, :, :].unsqueeze(0)
        image_grid = make_grid(channel_images, nrow=1)
        
        ax = axes[row, col]
        ax.imshow(image_grid.permute(1, 2, 0).squeeze(), cmap='gray')
        ax.axis('off')

    plt.subplots_adjust(wspace=0.05, hspace=0.05)
    plt.show()

# Example usage:
show_tensor_images(fake_dcgan, num_images=9, num_channels=3, num_columns=3)
show_tensor_images(fake_wgangp, num_images=9, num_channels=3, num_columns=3)
```
