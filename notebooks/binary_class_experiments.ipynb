```python
from torchvision.utils import make_grid
def show_tensor_images(image_tensor, num_images=64, size=(3, 28, 28), save=False, name=None, epoch=None):
    image_tensor = (image_tensor + 1) / 2
    image_unflat = image_tensor.detach().cpu()

    # Iterate through each channel
    for channel in range(size[0]):
        channel_images = image_unflat[:, channel:channel+1, :, :]
        image_grid = make_grid(channel_images[:num_images], nrow=8)
        
        plt.figure()
        plt.imshow(image_grid.permute(1, 2, 0).squeeze(), cmap='gray')
        plt.axis('off')
        if save:
            plt.savefig('{}_channel_{}_epoch_{}.png'.format(name, channel, epoch))
        plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal

# Set mean and covariance for the Gaussian distribution
mean = [14, 14]
#generate a numbre between 5 and 15
cov=np.random.uniform(1,3)
covariance = [[cov, 0], [0, cov]]  # Adjust covariance matrix as needed

# Create a grid
x, y = np.meshgrid(np.arange(28), np.arange(28))

# Create a multivariate normal distribution
gaussian = multivariate_normal(mean=mean, cov=covariance)

# Evaluate the Gaussian at each point in the grid
z = gaussian.pdf(np.dstack((x, y)))

# Normalize
z = z / z.max()

# Display the 2D Gaussian using imshow
plt.imshow(z, cmap='viridis', extent=(0, 28, 0, 28), origin='lower')
plt.title('SN source image')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.colorbar()
plt.show()
```
